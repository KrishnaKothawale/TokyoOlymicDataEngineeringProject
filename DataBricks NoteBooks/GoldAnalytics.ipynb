```python
athletes_clean_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Transformed Data/Athletes"
coaches_clean_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Transformed Data/Coaches"
teams_clean_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Transformed Data/Teams"
entriesGender_clean_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Transformed Data/EntriesGender"
medals_clean_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Transformed Data/Medals"
```

```python
athletes_clean = spark.read.format('csv').option("header","true").option("inferSchema","true").load(athletes_clean_path)

coaches_clean = spark.read.format('csv').option("header","true").option("inferSchema","true").load(coaches_clean_path)

medals_clean = spark.read.format('csv').option("header","true").option("inferSchema","true").load(medals_clean_path)

teams_clean = spark.read.format('csv').option("header","true").option("inferSchema","true").load(teams_clean_path)

entriesGender_clean = spark.read.format('csv').option("header","true").option("inferSchema","true").load(entriesGender_clean_path)
```

### **Top countries with respect to number of medals (rank)**

```python
medals_clean.show()
```

```python
from pyspark.sql.functions import col,isnan,when,count

top_countries = medals_clean.orderBy("Total", ascending=False)

top_countries.show()
```

### **Top Sports By Athlete Count**

```python
top_sports = athletes_clean.groupBy("Discipline").agg(count("*").alias("Athlete_count")).orderBy("Athlete_count", ascending = False)

top_sports.show()
```

### **Coaches Per Country**

```python
coach_analysis = coaches_clean.groupBy(
    "NOC"
).count().orderBy("count", ascending=False)

coach_analysis.show()
```

### **Women Percentage Per Sport**

```python
distribution = entriesGender_clean.select("Discipline", (col('Avg_Female') * 100).alias("Female_Percentage"), (col('Avg_Male') * 100).alias("Male_Percentage"))

distribution.show()
```

### **Overview**

```python
total_athletes = athletes_clean.count()

total_disciplines = athletes_clean.select(
    "Discipline"
).distinct().count()

total_countries = athletes_clean.select(
    "NOC"
).distinct().count()

total_medals = medals_clean.agg(
    {"Total":"sum"}
).collect()[0][0]
```

```python
overview = spark.createDataFrame(
    [
        (
            total_athletes,
            total_disciplines,
            total_countries,
            total_medals
        )
    ],
    [
        "Total_Athletes",
        "Total_Disciplines",
        "Total_Countries",
        "Total_Medals"
    ]
)

overview.show()
```

```python
top_countries.write.mode("overwrite").option("header", "true").csv("/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/country_medal_summary")
top_sports.write.mode("overwrite").option("header", "true").csv("/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/sports_popularity")
coach_analysis.write.mode("overwrite").option("header", "true").csv("/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/coach_distribution")
distribution.write.mode("overwrite").option("header", "true").csv("/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/gender_distribution")
```

```python
overview.coalesce(1).write.mode("overwrite").option("header", "true").csv("/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/overview")
```
