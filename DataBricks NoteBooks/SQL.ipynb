```python
country_medal_summary_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/country_medal_summary"
sports_popularity_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/sports_popularity"
coach_distribution_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/coach_distribution"
gender_distribution_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/gender_distribution"
overview_path = "/Volumes/tokyoolympicsdb2/tokyoolympicschema/tokyo-olympic-data-volume/Analytics/overview"
```

```python
top_countries = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(country_medal_summary_path)

top_sports = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(sports_popularity_path)

coach_analysis = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(coach_distribution_path)

distribution = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(gender_distribution_path)

overview = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv(overview_path)
```

```python
top_sports.printSchema()
```

```python
top_countries.write.mode("overwrite").saveAsTable("tokyoolympicsdb2.tokyoolympicschema.country_medal_summary")
```

```python
top_sports.write.mode("overwrite").saveAsTable("tokyoolympicsdb2.tokyoolympicschema.sport_popularity")

distribution.write.mode("overwrite").saveAsTable("tokyoolympicsdb2.tokyoolympicschema.gender_distribution")

coach_analysis.write.mode("overwrite").saveAsTable("tokyoolympicsdb2.tokyoolympicschema.coach_distribution")

overview.write.mode("overwrite").saveAsTable("tokyoolympicsdb2.tokyoolympicschema.overview")
```
